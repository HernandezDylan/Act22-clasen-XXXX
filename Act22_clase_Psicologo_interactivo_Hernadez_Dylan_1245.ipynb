#### Psicologo

```python
class Psicologo:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    def datos(self):
        print(self.nombre, self.telefono, self.edad)

    def escuchar(self, paciente):
        print(self.nombre, "escucha a", paciente)

    def diagnosticar(self, paciente):
        print(self.nombre, "diagnostica a", paciente)

    def terapia(self, paciente):
        print(paciente, "recibe terapia")

print("=== Psicologo ===")

doctor = Psicologo(
    input("Nombre: "),
    input("Teléfono: "),
    input("Edad: ")
)

while True:

    print("\n1.Datos 2.Escuchar 3.Diagnosticar 4.Terapia 5.Salir")
    op = input("Opción: ")

    if op == "1":
        doctor.datos()

    elif op == "2":
        doctor.escuchar(input("Paciente: "))

    elif op == "3":
        doctor.diagnosticar(input("Paciente: "))

    elif op == "4":
        doctor.terapia(input("Paciente: "))

    elif op == "5":
        print("Programa realizado por Hernandez Dylan 1245")
        break

    else:
        print("Error")
```
