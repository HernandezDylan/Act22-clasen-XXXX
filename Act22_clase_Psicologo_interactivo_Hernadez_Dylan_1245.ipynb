#### Psicologo
    
    

```python
class Psicologo:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Ahora las funciones reciben el nombre del paciente como argumento
    def escuchar(self, paciente):
        print(f"\n[Sesión] El Dr. {self.nombre} está escuchando atentamente a: {paciente}.")

    def diagnosticar(self, paciente):
        print(f"\n[Análisis] El Dr. {self.nombre} está redactando el diagnóstico de: {paciente}.")

    def dar_terapia(self, paciente):
        print(f"\n[Terapia] {paciente} está recibiendo terapia con el Dr. {self.nombre}.")

# --- Inicio del Programa ---
print("--- REGISTRO DE PSICÓLOGO ---")
nom_dr = input("Nombre del psicólogo: ")
tel_dr = input("Teléfono: ")
edad_dr = input("Edad: ")

# Creamos al psicólogo
doctor = Psicologo(nom_dr, tel_dr, edad_dr)

print(f"\n--- CONSULTORIO DEL DR. {doctor.nombre} ---")

while True:
    print("\nOpciones: 1. Atender Paciente | 2. Salir")
    menu = input("Seleccione: ")

    if menu == "1":
        paciente_nombre = input("¿Cómo se llama el paciente?: ")

        print(f"\n¿Qué acción realizará con {paciente_nombre}?")
        print("1. Escuchar | 2. Diagnosticar | 3. Dar Terapia")
        accion = input("Seleccione (1-3): ")

        if accion == "1": doctor.escuchar(paciente_nombre)
        elif accion == "2": doctor.diagnosticar(paciente_nombre)
        elif accion == "3": doctor.dar_terapia(paciente_nombre)
        elif accion == "4": print("Programa realizado por Hernadez Dylan 1245")
        else: print("Acción no válida.")

    elif menu == "2":
        print("Cerrando sistema médico...")
        break
    else:
        print("Opción no válida.")
```
